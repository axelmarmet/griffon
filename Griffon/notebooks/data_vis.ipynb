```python
from glob import glob
import pickle

import matplotlib.pyplot as plt

import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout

from griffon.coq_dataclasses import *
```

```python
files = glob("../data/intermediary/for_vis/**/*.pickle")

def insert_newlines(string, every=64):
    return '\n'.join(string[i:i+every] for i in range(0, len(string), every))

def draw_sample(sample:Stage1Statement, ax):
    graph = sample.ast
    # pos = graphviz_layout(graph, prog="dot")
    pos = graphviz_layout(graph, prog="dot")

    nx.draw(graph, pos, ax = ax)
    labels = {node_id : token for token, node_id in zip(sample.tokens, sample.token_to_node.values())}
    nx.draw_networkx_labels(graph, pos, labels, ax=ax)

    title = " ".join(sample.tokens)
    title = insert_newlines(title)
    ax.title.set_text(title)

_, axs = plt.subplots(5,5, figsize=(160,160))
for y in range(5):
    for x in range(5):
        if y*5 + x > 22:
            continue

        sample = pickle.load(open(files[y*5 + x], "rb"))
        draw_sample(sample.goal, axs[y,x])
```
